```python
from sklearn.datasets import make_blobs
#make blobs is used for generating synthetic data generator , used to generate usefull data for clustering , classification algorithms 
```

```python
import pandas as  pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns  
```

```python
#creating datasets 

#creating the normal datasets 
X_normal , y_normal = make_blobs(n_samples = 500 , centers = 3 )
```

```python
fig,  (ax1 , ax2)  = plt.subplots(1 , 2 , figsize = (12 , 5 ))
ax1.hist(X_normal.ravel() , bins = 30 )
ax2.hist(y_normal.ravel() , bins = int((len(y_normal) * ((30 * 100 / (len(X_normal.ravel())) )))/100 )  , color = 'red'  , alpha = 0.6)
plt.show()
```

```python
plt.scatter(X_normal[: , 0]  , X_normal[: , 1 ])
```
